```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Paramtres du problme
total_black = 50
total_white = 50

# Grilles de valeurs pour b1 et w1
b1_vals = np.arange(0, total_black + 1)
w1_vals = np.arange(0, total_white + 1)
B1, W1 = np.meshgrid(b1_vals, w1_vals)

# Fonction de probabilit
def compute_probability(b1, w1):
    n1 = b1 + w1
    if n1 == 0 or (100 - n1) == 0:
        return np.nan  # ignorer les cas o un sac est vide
    b2 = total_black - b1
    w2 = total_white - w1
    p1 = b1 / n1
    p2 = b2 / (b2 + w2)
    return 0.5 * (p1 + p2)

# Calcul de la surface de probabilit
P = np.vectorize(compute_probability)(B1, W1)

# Trac
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(B1, W1, P, cmap='viridis')

ax.set_xlabel('b1 (boules noires dans S1)')
ax.set_ylabel('w1 (boules blanches dans S1)')
ax.set_zlabel('Probabilit de tirer une noire')
ax.set_title('Surface de probabilit (jeu des 50 blanches / 50 noires)')

plt.tight_layout()
plt.show()
```
